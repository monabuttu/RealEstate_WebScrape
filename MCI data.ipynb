```python
import json
import requests
import numpy as np
import pandas as pd
```

```python
response = requests.get("https://services.arcgis.com/S9th0jAJ7bqgIRjw/arcgis/rest/services/MCI_2014_to_2018/FeatureServer/0/query?where=1%3D1&outFields=*&geometry=-79.473%2C43.723%2C-79.419%2C43.733&geometryType=esriGeometryEnvelope&inSR=4326&spatialRel=esriSpatialRelIntersects&outSR=4326&f=json")
```

```python
reponse_json=response.json()
reponse_json
```

```python
reponse_json.keys()
```

```python
len(reponse_json["features"])
```

```python
lat_l = []
lon_l = []
Neighbourhood_l=[]
Neighbourhood_ID=[]
MCI_l=[]
Year_l=[]
Month_l=[]
Day_l=[]


for x in np.arange(len(reponse_json["features"])):
    lat_l.append (reponse_json["features"][x]["attributes"]["Lat"])
    lon_l.append(reponse_json["features"][x]["attributes"]["Long"])
    Neighbourhood_l.append(reponse_json["features"][x]["attributes"]["Neighbourhood"])
    Neighbourhood_ID.append (reponse_json["features"][x]["attributes"]["Hood_ID"])
    MCI_l.append (reponse_json["features"][x]["attributes"]["MCI"])
    Year_l.append (reponse_json["features"][x]["attributes"]["reportedyear"])
    Month_l.append (reponse_json["features"][x]["attributes"]["reportedmonth"])
    Day_l.append (reponse_json["features"][x]["attributes"]["reportedday"])

```

```python
dataframe_dict = {"Latitude":lat_l,
                 "Longitude":lon_l,
                 "Neighbourhood":Neighbourhood_l,
                "Neighbourhood ID":Neighbourhood_ID,
                 "MCI":MCI_l,
                "Year":Year_l,
                "Month":Month_l,
                "Day":Day_l}
```

```python
df = pd.DataFrame(dataframe_dict)
df.head(20)
```
