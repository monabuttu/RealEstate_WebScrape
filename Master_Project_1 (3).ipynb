```python
import requests 
import pandas as pd
import numpy as np
import re
import os
```

```python
maindf = pd.DataFrame({"Address":''},index=[np.arange(1,20)])
maindf["Latitude"]=""
maindf["Longitude"]=""
maindf["Property Type"]=""
maindf["Lease Price"]=""
maindf["Sale Price"]= ""    
maindf['SizeInterior']=""
maindf["Total Size"]=""
maindf.head()
```

```python
Querycounter = 1
responses = []
x=0

while x<3:
    
    for index, row in maindf.iterrows(): 
        x+=1
        try:
            
            latmin = 43.6406
            latmax = 43.736439
            lngmin = -79.252357       
            lngmax = -79.549301


            url = 'https://api2.realtor.ca/Listing.svc/PropertySearch_Post'
            body = {"CultureId":1,
            "ApplicationId":1,
            "PropertySearchTypeId":1,
            "TransactionTypeID":2,
            "BuildingTypeId":0,
            "LongitudeMin":f'{lngmin}',
            "LongitudeMax":f'{lngmax}',
            "LatitudeMin":f'{latmin}',
            "LatitudeMax":f'{latmax}',
            "PriceMin":100000,
            "RecordsPerPage":10,
            "CurrentPage":[x]}

            response = requests.post(url, data=body).json()
            responses.append(response)
            print(f'Query {Querycounter}')
            Querycounter+=1

            if (response['Results'][0]['Property']['Address']['Longitude'] < str(lngmin)) and (response['Results'][0]['Property']['Address']['Longitude'] > str(lngmax)):
                continue
            else:
                pass
            
            if response['Results'][0]['Property']['Address']['AddressText'] in maindf['Address'].unique():
                continue
            else:
                maindf.loc[index, "Latitude"] = response['Results'][0]['Property']['Address']['Latitude']  
                maindf.loc[index, "Longitude"] = response['Results'][0]['Property']['Address']['Longitude']
                maindf.loc[index,"Address"] = response['Results'][0]['Property']['Address']['AddressText']        

            maindf.loc[index,"Property Type"] = response['Results'][0]['Property']['Type']

        except:
            print('Error')    

        try:
            maindf.loc[index,"Total Size"] = response['Results'][0]['Land']['SizeTotal']

        except:   
            maindf.loc[index,"Total Size"] = ""    

        try:
            maindf.loc[index,"Sale Price"] = response['Results'][0]['Property']['Price']
        except:
            maindf.loc[index,"Sale Price"] = ""

        try:
            print(response['Results'][0]['Building']['SizeInterior'])
            maindf.loc[index,"Sq. Ftg."] = response['Results'][0]['Building']['SizeInterior']
        except:
            print('Size invalid')
            maindf.loc[index,"Sq. Ftg."] = ""
```
