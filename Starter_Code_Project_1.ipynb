```python
import requests 
import pandas as pd
import numpy as np

url = 'https://api2.realtor.ca/Listing.svc/PropertySearch_Post'
body = {"CultureId":1,
"ApplicationId":1,
"PropertySearchTypeId":1,
"LongitudeMin":-43.6532,
"LongitudeMax":-79.3832,
"LatitudeMin":43.038275,
"LatitudeMax":43.044275}
response = requests.post(url, data=body).json()
```
