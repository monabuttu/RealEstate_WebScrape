```python
import requests 
import pandas as pd
import numpy as np
import re
```

```python
maindf = pd.DataFrame({"Address":''},index=[np.arange(1,50)])
maindf["Latitude"]=""
maindf["Longitude"]=""
maindf["Property Type"]=""
maindf["Lease Price"]=""
maindf["Sale Price"]= ""    
maindf['Ownership Type']=""
maindf["Total Size"]=""
maindf.head()
```

```python
Querycounter = 1
responses = []
x=0

while x<6:
    
    for index, row in maindf.iterrows(): 
        x+=1
        try:
            latmin = np.random.uniform(low=40.3, high=43.6)
            latmax = np.random.uniform(low=43.6, high=43.9)

            lngmin = np.random.uniform(low=-75.6, high=-79.9)            
            lngmax = np.random.uniform(low=-79.3, high=-79.6)


            url = 'https://api2.realtor.ca/Listing.svc/PropertySearch_Post'
            body = {"CultureId":1,
            "ApplicationId":1,
            "PropertySearchTypeId":1,
            "TransactionTypeID":2,
            "BuildingTypeId":0,
            "LongitudeMin":{lngmin},
            "LongitudeMax":{lngmax},
            "LatitudeMin":{latmin},
            "LatitudeMax":{latmax},
            "PriceMin":100000,
            "RecordsPerPage":10,
            "CurrentPage":[x]}

            response = requests.post(url, data=body).json()
            responses.append(response)
            print(f'Query {Querycounter}')
            Querycounter+=1

            if response['Results'][0]['Property']['Address']['AddressText'] in maindf['Address'].unique():
                continue
            else:
                maindf.loc[index, "Latitude"] = response['Results'][0]['Property']['Address']['Latitude']  
                maindf.loc[index, "Longitude"] = response['Results'][0]['Property']['Address']['Longitude']
                maindf.loc[index,"Address"] = response['Results'][0]['Property']['Address']['AddressText']        
                maindf.loc[index,"Property Type"] = response['Results'][0]['Property']['Type']
                maindf.loc[index,"Ownership Type"] = response['Results'][0]['Property']['OwnershipType']
        except:
            print('Error')    

        try:
            maindf.loc[index,"Total Size"] = response['Results'][0]['Land']['SizeTotal']

        except:   
            maindf.loc[index,"Total Size"] = ""    

        try:
            maindf.loc[index,"Sale Price"] = response['Results'][0]['Property']['Price']
        except:
            maindf.loc[index,"Sale Price"] = ""
```
