```python
import requests 
import pandas as pd
import numpy as np
```

```python
maindf = pd.DataFrame({"Address":''},index=[np.arange(1,50)])
maindf["Latitude"]=""
maindf["Longitude"]=""
maindf["Property Type"]=""
maindf["Building Type"]=""
maindf["Construction Type"]=""
maindf["Sale Price"]= ""    
maindf['Size Interior']=""
maindf['Postal Code']=""
maindf.head()
```

```python
Querycounter = 1
responses = []
for index, row in maindf.iterrows(): 
    try:
        latmin = np.random.uniform(low=41.3, high=43.6)
        latmax = np.random.uniform(low=43.6, high=43.9)
                                   
        lngmin = np.random.uniform(low=-76.6, high=-79.9)            
        lngmax = np.random.uniform(low=-79.3, high=-79.6)
        
        
        url = 'https://api2.realtor.ca/Listing.svc/PropertySearch_Post'
        body = {"CultureId":1,
        "ApplicationId":1,
        "PropertySearchTypeId":1,
        "TransactionTypeID":2,
        "BuildingTypeId":0,
        "LongitudeMin":{lngmin},
        "LongitudeMax":{lngmax},
        "LatitudeMin":{latmin},
        "LatitudeMax":{latmax}}
        response = requests.post(url, data=body).json()
        responses.append(response)
        print(f'Query {Querycounter}')
        Querycounter+=1
        
        maindf.loc[index, "Latitude"] = response['Results'][0]['Property']['Address']['Latitude']
        maindf.loc[index, "Longitude"] = response['Results'][0]['Property']['Address']['Longitude']
        maindf.loc[index,"Address"] = response['Results'][0]['Property']['Address']['AddressText'] 
        maindf.loc[index,"Property Type"] = response['Results'][0]['Property']['Type']
        maindf.loc[index,"Building Type"] = response['Results'][0]['Building']['Type']
        maindf.loc[index,"Construction Type"] = response['Results'][0]['Property']['OwnershipType']
        maindf.loc[index,"Sale Price"] = response['Results'][0]['Property']['Price']
        #df.loc[index,""] = response['Results'][0]['Land']['SizeTotal']
        maindf.loc[index,"Size Interior"] = response['Results'][0]['Building']['SizeInterior']
        maindf.loc[index,"Postal Code"] = response['Results'][0]['PostalCode']
        print('---------------------------------------------------------------------------------------------------------------')
    except:
        print(f'City for coordinates')
```
